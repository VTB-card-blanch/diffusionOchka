```python
import gc
from diffusers import AutoencoderKL, StableDiffusionXLPipeline # type: ignore
from PIL import Image # type: ignore
from safetensors.torch import load_file # type: ignore
import torch # type: ignore
from torchvision.transforms import v2 # type: ignore

def flush():
    gc.collect()
    torch.cuda.empty_cache()


device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Используемое устройство: {device}")
```

```python
print("Загрузка базового SDXL...")

# Загружаем пайплайн
pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant="fp16",
    low_cpu_mem_usage=True,
).to(device)
```

```python
# Промпт для теста
prompt = "A cinematic shot of a modern android on the street, rain on the streets, high detail, 8k"
#prompt = "Кинематографический снимок современного андроида на улице, дождь на улицах, высокая детализация, 8k"
#prompt = "Нарисуй вазу и цветы"

# Генерация
image_base = pipe(prompt=prompt).images[0]
image_base.save("sdxl_base.png")
```
